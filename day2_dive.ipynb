```python
import re
from collections import defaultdict
```

```python
filename = 'day2_dive.txt'
```

```python
with open(filename, 'r') as infile:
    commands = infile.readlines()
```

# Part 1

```python
parsed = [re.match(r'(\w+) (\d+).*', cmd).groups() for cmd in commands]
```

```python
totals = defaultdict(int)
```

```python
for cmd, val in parsed:
    totals[cmd] += int(val)
```

```python
totals
```

```python
fwd = totals['forward']
depth = totals['down'] - totals['up']
print(fwd)
print(depth)
```

```python
fwd*depth
```

# Part 2

```python
class State:
    def __init__(self):
        self.fwd = 0
        self.depth = 0
        self.aim = 0
        
    def run_cmd(self, command, val):
        if command == 'up':
            #self.depth -= val
            self.aim -= val
        elif command == 'down':
            #self.depth += val
            self.aim += val
        elif command == 'forward':
            self.fwd += val
            self.depth += val*self.aim
            
    def __str__(self):
        return f'fwd = {self.fwd}, depth = {self.depth}, aim = {self.aim}'
```

```python
s = State()

for cmd, val in parsed:
    s.run_cmd(cmd, int(val))
    print(cmd, val)
    print(s)
```

```python
s.fwd * s.depth
```
